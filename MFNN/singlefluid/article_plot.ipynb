```python
# 神经网络的类
class PINN_NeuralNet(nn.Module):
    """ Set basic architecture of the PINN model."""

    def __init__(self,
                 input_dim=0,
                 output_dim=1,  # 默认输出维度为1
                 num_hidden_layers=4, 
                 num_neurons_per_layer=20,
                 activation='tanh',
                 kernel_initializer='glorot_normal',
                 **kwargs):
        super(PINN_NeuralNet, self).__init__()

        self.num_hidden_layers = num_hidden_layers
        self.output_dim = output_dim
        
        # 添加输入层
        self.input_layer = nn.Linear(input_dim, num_neurons_per_layer)
        
        # 添加其他隐藏层
        self.hidden_layers = nn.ModuleList()
        for _ in range(num_hidden_layers):
            self.hidden_layers.append(nn.Linear(num_neurons_per_layer, num_neurons_per_layer))
        
        # 添加输出层
        self.out = nn.Linear(num_neurons_per_layer, output_dim)
        # 设置激活函数
        if activation == 'tanh':
            self.activation = torch.tanh
        elif activation == 'relu':
            self.activation = F.relu6
        elif activation == 'sigmoid':
            self.activation = torch.sigmoid
        elif activation == 'linear':
            self.activation = None
        else:
            raise ValueError("Unsupported activation function")
        
        # 初始化权重
        if kernel_initializer == 'glorot_normal':
            nn.init.xavier_normal_(self.input_layer.weight)
            for hidden_layer in self.hidden_layers:
                nn.init.xavier_normal_(hidden_layer.weight)
            nn.init.xavier_normal_(self.out.weight)
        elif kernel_initializer == 'glorot_uniform':
            nn.init.xavier_uniform_(self.input_layer.weight)
            for hidden_layer in self.hidden_layers:
                nn.init.xavier_uniform_(hidden_layer.weight)
            nn.init.xavier_uniform_(self.out.weight)
        else:
            raise ValueError("Unsupported kernel initializer")

    def forward(self, X):

        # 进入输入层
        Z = self.input_layer(X)
        
        # 通过隐藏层
        for hidden_layer in self.hidden_layers:
            Z = hidden_layer(Z)
            if self.activation is not None:
                Z = self.activation(Z)
        # 通过输出层输出
        Z = self.out(Z)
        
        return Z
```

```python
model_HF_nl_pinn = PINN_NeuralNet(input_dim=in_dim,
                             output_dim=out_dim,
                             num_hidden_layers=4,
                             num_neurons_per_layer=64,
                             activation='relu')
model_HF_l_pinn = PINN_NeuralNet(input_dim=in_dim,
                            output_dim=out_dim,
                            num_hidden_layers=1,
                            num_neurons_per_layer=10,
                            activation='linear')
model_HF_nl_dnn = PINN_NeuralNet(input_dim=in_dim,
                             output_dim=out_dim,
                             num_hidden_layers=4,
                             num_neurons_per_layer=64,
                             activation='relu')
model_HF_l_dnn = PINN_NeuralNet(input_dim=in_dim,
                            output_dim=out_dim,
                            num_hidden_layers=1,
                            num_neurons_per_layer=10,
                            activation='linear')
```

```python
def plot_residuals(model_HF_nl, model_HF_l, color, label, ax, test_data, lb, ub):
    """绘制残差图（频率和真实值均取对数）"""
    # 获取测试数据的频率和真实值
    x1_test = np.log10(test_data['AngFreq'].values)  # 频率取对数
    y_true = np.log10(test_data['StorageM'].values)  # 真实值取对数

    # 对测试数据的频率点进行预测
    DP_plot = norm(162. * np.ones_like(x1_test).reshape(-1, 1), lb[0], ub[0])  # DP 值归一化
    Mn_plot = norm(np.unique(test_data['Mn']) * np.ones_like(x1_test).reshape(-1, 1), lb[1], ub[1])  # Mn 值归一化
    PDI_plot = norm(np.unique(test_data['PDI']) * np.ones_like(x1_test).reshape(-1, 1), lb[2], ub[2])  # PDI 值归一化
    w_plot = norm(x1_test.reshape(-1, 1), lb[3], ub[3])  # 频率（已取对数）归一化

    # 构建输入数据
    X_MF_np = np.hstack((DP_plot, Mn_plot, PDI_plot, w_plot))
    X_MF = torch.tensor(X_MF_np, dtype=torch.float32)

    # 模型预测
    y_MF = model_HF_nl(X_MF) + model_HF_l(X_MF)
    y_pred = denorm(y_MF.detach().numpy(), lb[4], ub[4])  # 反归一化
    y_pred = np.log10(10 ** y_pred.squeeze())  # 取对数

    # 计算残差
    residuals = y_true - y_pred

    # 绘制残差图
    ax.scatter(x1_test, residuals, color=color, label=label,s=50)  # 频率反取对数后绘制

# 获取测试数据
DP_range = [162.]
test_data = df_hf[df_hf.DP == DP_range[0]]
x1_test = test_data['AngFreq'].values  # 未对数化、未归一化, np
y1_test = test_data['StorageM'].values  # 未对数化，未归一化, np

# 创建残差图的图形和坐标轴
fig_res, ax_res = plt.subplots(figsize=(8, 6), dpi=600)

# 绘制DNN模型的残差图
plot_residuals(model_HF_nl_dnn, model_HF_l_dnn, 'tab:blue', 'DNN', ax_res, test_data, lb, ub)

# 绘制PINN模型的残差图
plot_residuals(model_HF_nl_pinn, model_HF_l_pinn, 'tab:green', 'PINN', ax_res, test_data, lb, ub)

# 绘制 y=0 的红色虚线
ax_res.axhline(0, color='red', linestyle='--', lw=3)

# 设置残差图的属性
ax_res.set_ylabel('lg(Residuals) G\'')
ax_res.set_xlabel('$lg\omega$')
ax_res.legend(fontsize=18, loc='upper left', frameon=False)

# 显示残差图
plt.show()
```

```python
def plot_residuals(model_HF_nl, model_HF_l, color, label, ax, test_data, lb, ub):
    """绘制 tanδ 的频率残差图（横坐标是频率的对数，纵坐标是残差的对数）"""
    # 获取测试数据的频率和真实值
    x1_test = np.log10(test_data['AngFreq'].values)  # 频率取对数
    y_true = test_data['LossFactor'].values # 真实值取对数

    # 对测试数据的频率点进行预测
    DP_plot = norm(162. * np.ones_like(x1_test).reshape(-1, 1), lb[0], ub[0])  # DP 值归一化
    Mn_plot = norm(np.unique(test_data['Mn']) * np.ones_like(x1_test).reshape(-1, 1), lb[1], ub[1])  # Mn 值归一化
    PDI_plot = norm(np.unique(test_data['PDI']) * np.ones_like(x1_test).reshape(-1, 1), lb[2], ub[2])  # PDI 值归一化
    w_plot = norm(x1_test.reshape(-1, 1), lb[3], ub[3])  # 频率（已取对数）归一化

    # 构建输入数据
    X_MF_np = np.hstack((DP_plot, Mn_plot, PDI_plot, w_plot))
    X_MF = torch.tensor(X_MF_np, dtype=torch.float32)

    # 模型预测
    y_MF = model_HF_nl(X_MF) + model_HF_l(X_MF)
    y_pred = denorm(y_MF.detach().numpy(), lb[4], ub[4])  # 反归一化
    y_pred = y_pred.flatten()

    # 计算残差
    residuals = y_true - y_pred

    # 绘制残差图
    ax.scatter(x1_test, residuals, color=color, label=label, s=50)  # 横坐标为频率的对数值

    # 设置 y 轴范围（根据 residuals 的实际范围调整）
    residuals_min = min(np.min(residuals), -0.4)
    residuals_max = max(np.max(residuals), 0.4)
    ax.set_ylim(residuals_min, residuals_max)

# 创建残差图的图形和坐标轴
fig_res, ax_res = plt.subplots(figsize=(8, 6), dpi=600)

# 绘制DNN模型的残差图
plot_residuals(model_HF_nl_dnn, model_HF_l_dnn, 'tab:blue', 'DNN', ax_res, test_data, lb, ub)

# 绘制PINN模型的残差图
plot_residuals(model_HF_nl_pinn, model_HF_l_pinn, 'tab:green', 'PINN', ax_res, test_data, lb, ub)

# 绘制 y=0 的红色虚线
ax_res.axhline(0, color='red', linestyle='--', lw=3)

# 设置残差图的属性
ax_res.set_ylabel('lg(Residuals) tanδ')
ax_res.set_xlabel('$lg\omega$')
ax_res.legend(fontsize=18, loc='upper left', frameon=False)

# 显示残差图
plt.show()
```
